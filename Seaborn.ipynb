```python
import seaborn as sns
import matplotlib.pyplot as plt
fmri=sns.load_dataset("fmri")
sns.lineplot(x="timepoint",y="signal",hue="event",style="region",data=fmri)
plt.show()
```

# Qus 3

```python
titanic=sns.load_dataset("titanic")
plt.subplot(1,2,1)
sns.boxplot(x="pclass",y="age",data=titanic)
plt.subplot(1,2,2)
sns.boxplot(x="pclass",y="fare",data=titanic)
plt.tight_layout()
plt.show()
```

# Qus 4

```python
diamond=sns.load_dataset("diamonds")
sns.histplot(data=diamond,x="price",hue="cut",kde=True,palette="viridis",alpha=0.7)
plt.show()
```

# Qus 5

```python
iris=sns.load_dataset("iris")
sns.pairplot(iris,hue="species",palette="viridis")
plt.show()
```

# Qus 6

```python
flights=sns.load_dataset("flights")
pivot_flights=flights.pivot_table(index="month",columns="year",values="passengers")
sns.heatmap(pivot_flights,cmap="Y1GnBu",annot=True,fmt="d")
plt.show()
```
